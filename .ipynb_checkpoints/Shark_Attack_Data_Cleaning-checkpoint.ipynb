# conclusion 

## From 1423 attacks, we can see in the following bar charts

```python
df['fatal (y/n)'].value_counts().sum()
```

```python
from matplotlib.pyplot import figure
fig = figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
ax = fig.add_axes([0,0,1,1])
Activities = ['Swimming', 'Wading', 'Spearfishing', 'Surfing', 'Fishing']
Attacks = [len(swimming),len(wading),len(spearfishing),len(surfing),len(fish)]
ax.bar(Activities,Attacks)
plt.show()
```

## We can see that the Surfing is the most attacked activitie followed by Swimming, Spearfishing, Wading and fishing

```python
from matplotlib.pyplot import figure
fig = figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
ax = fig.add_axes([0,0,1,1])
Activities = ['Swimming', 'Wading', 'Spearfishing', 'Surfing', 'Fishing']
Deaths = [len(deathsswi),len(deathswading),len(deathsspear),len(deaths),len(deathsfish)]
ax.bar(Activities,Deaths)
plt.show()
```

## Analysing the data, Swimming is the riskiests activity, followed by Surfimg, Spearfishing and Wading. When people was fishing, we see there isn't any death.
